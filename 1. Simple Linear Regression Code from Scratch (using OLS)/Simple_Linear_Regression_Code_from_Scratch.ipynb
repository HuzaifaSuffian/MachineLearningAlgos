```python
# Importing the Data
df = pd.read_csv('placement.csv')
```

```python
# Spliting the data
X = df.iloc[:,0:1].values
y = df.iloc[:,-1].values
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# Fitting Sklearn Linear Model for Comparison
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# Fitting the data for training
lr.fit(X_train,y_train)
```

```python
# Predicting values for test data
lr.predict(X_test)
```

```python
# Finding the value of m (slope,coef_), b (y_intercept)
print('m:',lr.coef_)
print('b:',lr.intercept_)
```

```python
# Creating our own Simple Linear Regression class from scratch (using mathematical formulation)
class myLR:
  # Initializing values for m and b
  def __init__(self):
    self.m = None
    self.b = None

  def fit(self,X_train,y_train):
    num = 0
    den = 0

    # Loop for each data point
    for i in range(X_train.shape[0]):
      num = num + ((X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean()))
      den = den + ((X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean()))

    self.m = num/den
    self.b = y_train.mean() - self.m * X_train.mean()


  def predict(self,X_test):
    return self.m * X_test + self.b
```

```python
mylr = myLR()
```

```python
mylr.fit(X_train,y_train)
```

```python
# Doing Predict and converting the 2d numpy array to 1d
mylr.predict(X_test.ravel())
```
